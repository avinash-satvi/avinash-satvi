```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

def generate_synthetic_data(num_samples, num_features, num_clusters):
    data, labels = make_blobs(n_samples=num_samples, n_features=num_features, centers=num_clusters, random_state=42)
    return data, labels

def k_means_clustering(data, num_clusters):
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(data)
    centroids = kmeans.cluster_centers_
    return cluster_labels, centroids

def visualize_clusters(data, true_labels, centroids):
    plt.figure(figsize=(15, 10))

    # Scatter plot with true labels
    plt.subplot(2, 3, 1)
    plt.scatter(data[:, 0], data[:, 1], c=true_labels, cmap='viridis', alpha=0.5, edgecolors='w')
    plt.title('Scatter Plot with True Labels')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

    # Scatter plot of data points with cluster assignments
    plt.subplot(2, 3, 2)
    plt.scatter(data[:, 0], data[:, 1], c=true_labels, cmap='viridis', alpha=0.5, edgecolors='w')
    plt.title('Scatter Plot with True Labels')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

    # Scatter plot of data points with cluster assignments
    plt.subplot(2, 3, 3)
    plt.scatter(data[:, 0], data[:, 1], c=true_labels, cmap='viridis', alpha=0.5, edgecolors='w')
    plt.title('Scatter Plot with True Labels')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

    # Scatter plot with cluster assignments
    plt.subplot(2, 3, 4)
    plt.scatter(data[:, 0], data[:, 1], c=true_labels, cmap='viridis', alpha=0.5, edgecolors='w')
    plt.title('Scatter Plot with True Labels')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

    # Visualize cluster sizes
    plt.subplot(2, 3, 5)
    cluster_sizes = np.bincount(true_labels)
    plt.bar(range(len(cluster_sizes)), cluster_sizes, color='blue', alpha=0.7)
    plt.title('Cluster Sizes')
    plt.xlabel('Cluster')
    plt.ylabel('Number of Data Points')

    # Visualize cluster assignment changes
    plt.subplot(2, 3, 6)
    plt.scatter(range(len(true_labels)), true_labels, c=true_labels, cmap='viridis', marker='o')
    plt.title('Cluster Assignment Changes')
    plt.xlabel('Data Point Index')
    plt.ylabel('Cluster Assignment')

    plt.tight_layout()
    plt.show()

def main():
    # Parameters
    num_samples = 300
    num_features = 2
    num_clusters = 3  # You can change the number of clusters

    # Step 1: Generate Synthetic Dataset
    synthetic_data, true_labels = generate_synthetic_data(num_samples, num_features, num_clusters)

    # Step 2: K-Means Clustering
    predicted_labels, centroids = k_means_clustering(synthetic_data, num_clusters)

    # Step 3: Visualize Clusters
    visualize_clusters(synthetic_data, true_labels, centroids)

if __name__ == "__main__":
    main()
```
