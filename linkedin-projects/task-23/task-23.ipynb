```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.manifold import TSNE

# Generate larger synthetic dataset
X, y = make_blobs(n_samples=1000, centers=5, random_state=42, cluster_std=2.0)

# Apply t-SNE
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

# Plot the original and t-SNE-transformed data
plt.figure(figsize=(12, 5))

# Plot original data
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolors='k')
plt.title('Original Data')

# Plot t-SNE-transformed data
plt.subplot(1, 2, 2)
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', edgecolors='k')
plt.title('t-SNE Transformed Data')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Generate synthetic dataset
X, y = make_blobs(n_samples=1000, centers=5, random_state=42, cluster_std=2.0)

# Scatter Plot - Original Data
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', edgecolors='k')
plt.title('Scatter Plot - Original Data')
plt.show()
```

```python
# Color-Coding
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis', edgecolors='k')
plt.title('Scatter Plot - t-SNE Transformed Data (Color-Coded)')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from ipywidgets import interact
import matplotlib.pyplot as plt

# Generate or load your dataset, assuming X is your feature matrix and y is your labels
# Replace this with your actual data loading or generation logic
# Example:
# X, y = ...

# Standardize the features
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

# Scatter plot function
def scatter_plot(X, y, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Spectral)
    plt.title(title)
    plt.xlabel('t-SNE Dimension 1')
    plt.ylabel('t-SNE Dimension 2')
    plt.show()

# Interactive Visualizations
@interact(perplexity=(5, 50, 5))
def interactive_tsne(perplexity=30):
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)
    X_tsne_interactive = tsne.fit_transform(X_std)
    scatter_plot(X_tsne_interactive, y, f't-SNE Transformed Data (Perplexity={perplexity})')
```

```python
# Perplexity Parameter Exploration
perplexities = [5, 10, 20, 30, 40, 50]
plt.figure(figsize=(15, 8))
for i, perplexity in enumerate(perplexities):
    plt.subplot(2, 3, i + 1)
    tsne = TSNE(n_components=2, perplexity=perplexity, random_state=42)
    X_tsne_explore = tsne.fit_transform(X_std)
    plt.scatter(X_tsne_explore[:, 0], X_tsne_explore[:, 1], c=y, cmap='viridis', edgecolors='k')
    plt.title(f'Perplexity={perplexity}')
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import KMeans
# Label Propagation
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans_labels = kmeans.fit_predict(X_std)

scatter_plot(X_std, kmeans_labels, 'K-Means Clustering')
```

```python
from sklearn.metrics import silhouette_score

# Silhouette Analysis
silhouette_avg = silhouette_score(X_std, kmeans_labels)
print(f'Silhouette Score: {silhouette_avg}')
```
