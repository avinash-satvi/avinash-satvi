```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Generate a larger synthetic dataset
np.random.seed(42)

# Creating a 2D synthetic dataset with more data points
mean = [0, 0]
covariance = [[1, 0.8], [0.8, 1]]
data = np.random.multivariate_normal(mean, covariance, 1000)  # Increase the number of data points

# Visualize the original dataset
plt.scatter(data[:, 0], data[:, 1])
plt.title('Original Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# Applying PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(data)

# Visualize the PCA-transformed dataset
plt.scatter(pca_result[:, 0], pca_result[:, 1])
plt.title('PCA Transformed Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Explained Variance Ratio
explained_variance_ratio = pca.explained_variance_ratio_
print(f"Explained Variance Ratio: {explained_variance_ratio}")

# Cumulative Explained Variance
cumulative_explained_variance = np.cumsum(explained_variance_ratio)
print(f"Cumulative Explained Variance: {cumulative_explained_variance}")

# Plotting the explained variance
plt.plot(range(1, len(cumulative_explained_variance) + 1), cumulative_explained_variance, marker='o')
plt.title('Cumulative Explained Variance')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Visualize the original dataset with scatter plot
plt.scatter(data[:, 0], data[:, 1])
plt.title('Scatter Plot of Original Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
import seaborn as sns

# Create a biplot
sns.set(style="white")
plt.figure(figsize=(8, 8))

# Scatter plot of PCA components
plt.scatter(pca_result[:, 0], pca_result[:, 1], alpha=0.5)

# Plot the loading vectors
for i in range(pca.components_.shape[1]):
    plt.arrow(0, 0, pca.components_[0, i], pca.components_[1, i], color='r', alpha=0.8, linewidth=0.5, head_width=0.02)

plt.title('Biplot')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Plot the explained variance ratio
plt.bar(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio)
plt.title('Explained Variance Ratio')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

```python
# Heatmap of the covariance matrix
cov_matrix = np.cov(data.T)
sns.heatmap(cov_matrix, annot=True, fmt='g', cmap='coolwarm', xticklabels=['Feature 1', 'Feature 2'], yticklabels=['Feature 1', 'Feature 2'])
plt.title('Covariance Matrix Heatmap')
plt.show()
```

```python
# Individual factor map for the first two principal components
individual_factor_map = pca_result[:, :2]
plt.scatter(individual_factor_map[:, 0], individual_factor_map[:, 1])
plt.title('Individual Factor Map')
plt.xlabel('Individual Factor 1')
plt.ylabel('Individual Factor 2')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D

# Generate a larger synthetic 3D dataset
np.random.seed(42)
mean_3d = [0, 0, 0]
covariance_3d = [[1, 0.8, 0.5], [0.8, 1, 0.3], [0.5, 0.3, 1]]
data_3d = np.random.multivariate_normal(mean_3d, covariance_3d, 1000)

# Applying PCA for 3D data with maximum components of 3
n_components_3d = min(data_3d.shape[0], data_3d.shape[1])
pca_3d = PCA(n_components=n_components_3d)
pca_result_3d = pca_3d.fit_transform(data_3d)

# Visualize 3D scatter plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_result_3d[:, 0], pca_result_3d[:, 1], pca_result_3d[:, 2], c='r', marker='o')
ax.set_title('3D Scatter Plot')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.show()
```
