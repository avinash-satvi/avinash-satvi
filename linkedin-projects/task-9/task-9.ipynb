```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

def summarize_text(text, max_length=150, model_name="t5-small"):
    # Load pre-trained T5 model and tokenizer
    model = T5ForConditionalGeneration.from_pretrained(model_name)
    tokenizer = T5Tokenizer.from_pretrained(model_name)

    # Tokenize and generate the summary
    input_ids = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=512, truncation=True)
    summary_ids = model.generate(input_ids, max_length=max_length, length_penalty=2.0, num_beams=4, early_stopping=True)

    # Decode the summary
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return summary

# Example usage
input_text = """
    Transformers have revolutionized natural language processing. They use self-attention mechanisms 
    to process input sequences in parallel. One popular transformer model is T5, which stands for 
    Text-To-Text Transfer Transformer. Let's use T5 to generate a summary of this text.
"""
result = summarize_text(input_text)
print("Original Text:")
print(input_text)
print("\nSummarized Text:")
print(result)
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from rouge import Rouge
from textblob import TextBlob

# Function to generate word cloud
def generate_wordcloud(text, title):
    wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)
    
    # Use matplotlib directly for displaying the word cloud
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.title(title)
    plt.axis("off")
    plt.show()

# Function to generate bar chart for word frequencies
def visualize_word_frequencies(text, title, top_n=10):
    # Tokenize and count word frequencies
    tokens = text.split()
    word_freq = {word: tokens.count(word) for word in set(tokens)}
    top_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)[:top_n]

    # Create bar chart
    plt.figure(figsize=(10, 5))
    sns.barplot(x=[word[0] for word in top_words], y=[word[1] for word in top_words])
    plt.title(title)
    plt.xlabel("Word")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()

# Function to visualize sentence length distribution
def visualize_sentence_length_distribution(text, title):
    # Tokenize sentences and calculate lengths
    sentences = text.split('.')
    sentence_lengths = [len(sentence.split()) for sentence in sentences]
    
    # Plot distribution
    plt.figure(figsize=(10, 5))
    sns.histplot(sentence_lengths, kde=True)
    plt.title(title)
    plt.xlabel('Sentence Length')
    plt.ylabel('Frequency')
    plt.show()

# Function to visualize sentiment analysis
def visualize_sentiment_analysis(text, title):
    # Analyze sentiment using TextBlob
    blob = TextBlob(text)
    sentiment_scores = [sentence.sentiment.polarity for sentence in blob.sentences]
    
    # Plot sentiment distribution
    plt.figure(figsize=(10, 5))
    sns.histplot(sentiment_scores, kde=True)
    plt.title(title)
    plt.xlabel('Sentiment Polarity')
    plt.ylabel('Frequency')
    plt.show()

# Function to estimate reading time and visualize the comparison
def visualize_reading_time_comparison(original_text, summarized_text, words_per_minute=200):
    original_reading_time = len(original_text.split()) / words_per_minute
    summarized_reading_time = len(summarized_text.split()) / words_per_minute
    
    # Plot reading time comparison
    plt.figure(figsize=(10, 5))
    sns.barplot(x=['Original Text', 'Summarized Text'], y=[original_reading_time, summarized_reading_time])
    plt.title('Reading Time Comparison')
    plt.ylabel('Reading Time (minutes)')
    plt.show()

# Function to visualize attention heatmap
def visualize_attention_heatmap(input_text, summarized_text, model_name="t5-small"):
    # Load pre-trained T5 model and tokenizer
    model = T5ForConditionalGeneration.from_pretrained(model_name)
    tokenizer = T5Tokenizer.from_pretrained(model_name)

    # Tokenize and generate attention scores
    input_ids = tokenizer.encode("summarize: " + input_text, return_tensors="pt", max_length=512, truncation=True)
    outputs = model.encoder(input_ids)
    
    # Check if attention tensors are available
    if 'attentions' in outputs:
        attention = outputs['attentions']
        
        # Visualize attention scores using a heatmap
        plt.figure(figsize=(10, 5))
        sns.heatmap(np.mean(attention[0][0].detach().numpy(), axis=0), annot=True, cmap="viridis")
        plt.title("Attention Heatmap")
        plt.show()
    else:
        print("Attention tensors not available for visualization.")

# Example usage
input_text = """
    Transformers have revolutionized natural language processing. They use self-attention mechanisms 
    to process input sequences in parallel. One popular transformer model is T5, which stands for 
    Text-To-Text Transfer Transformer. Let's use T5 to generate a summary of this text.
"""
result = summarize_text(input_text)

# Visualizations
generate_wordcloud(input_text, "Word Cloud for Original Text")
generate_wordcloud(result, "Word Cloud for Summarized Text")

visualize_word_frequencies(input_text, "Top Words in Original Text")
visualize_word_frequencies(result, "Top Words in Summarized Text")

visualize_attention_heatmap(input_text, result)
visualize_rouge_scores(input_text, result)
visualize_sentence_length_distribution(input_text, 'Sentence Length Distribution in Original Text')
visualize_sentence_length_distribution(result, 'Sentence Length Distribution in Summarized Text')
visualize_sentiment_analysis(input_text, 'Sentiment Analysis in Original Text')
visualize_sentiment_analysis(result, 'Sentiment Analysis in Summarized Text')
visualize_reading_time_comparison(input_text, result)
```
